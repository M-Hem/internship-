```
import pandas as pd


df = pd.read_csv("/content/marvel_ratings.csv")

# 1. Display first 5 rows
print("1. Head:")
print(df.head())

# 2. Display last 5 rows
print("\n2. Tail:")
print(df.tail())

# 3. Display dataset information
print("\n3. Info:")
print(df.info())

# 4. Display statistical summary
print("\n4. Describe:")
print(df.describe())

# 5. Display column names
print("\n5. Columns:")
print(df.columns)

# 6. Display shape of the dataset
print("\n6. Shape:")
print(df.shape)

# 7. Check for missing values
print("\n7. Missing Values:")
print(df.isnull().sum())

# 8. Remove duplicate rows
print("\n8. Remove Duplicates:")
df_no_duplicates = df.drop_duplicates()
print(df_no_duplicates)

# 9. Sort data by the first column
print("\n9. Sorted Data:")
sorted_df = df.sort_values(by=df.columns[0])
print(sorted_df)

# 10. Display unique values of the first column
print("\n10. Unique Values in First Column:")
print(df[df.columns[0]].unique())
```
